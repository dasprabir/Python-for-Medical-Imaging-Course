```python
import pydicom
ds = pydicom.dcmread('Image3.dcm')
print(type(ds))
```

```python
# Check if "Clinical Trial Series ID" tag (0012,0071) is present
tag = (0x0012, 0x0071)
if tag in ds:
    print("✅ 'Clinical Trial Series ID' is present.")
    print("Value:", ds[tag].value)
else:
    print("❌ 'Clinical Trial Series ID' is NOT present.")
```

```python
import pydicom
ds = pydicom.dcmread('Image3.dcm')
tag = (0x0018,0x0081)

if tag in ds:
    vr = ds[tag].VR
else:
    print('VR tag NOT present')

print(vr)
```

```python
import pydicom

# Load the DICOM file
dicom_path = "Image3.dcm"
ds = pydicom.dcmread(dicom_path)

# Tag (0018,0081): Echo Time
tag = (0x0018, 0x0081)

if tag in ds:
    print(f"VR of tag (0018,0081): {ds[tag].VR}")
else:
    print("Tag (0018,0081) not found in the DICOM file.")
```
